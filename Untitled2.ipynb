## Exploratory data analysis and Data Visualization


 ### List of the countries present in the dataset for the survey

```python
print("List of the country in the dataset")
for x in pd.unique(zomato.Country):
    print(x)
```

```python
max_country=zomato.Country.value_counts().index
val_country=zomato.Country.value_counts().values
plt.pie(val_country[:3], labels=max_country[:3],labeldistance=1.15, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' })
```

### Conclusion drawn from the above visualization:
1: India has the maximum number of restaurants on Zomato.<br> 
2: Zomato generate major revenue from India, after USA.


```python
stopwords=set(STOPWORDS)
wordcloud=(WordCloud(width=500, height=500, relative_scaling=0.5,
stopwords=stopwords).generate_from_frequencies(zomato['Restaurant Name'].value_counts().head(40)))
fig=plt.figure(1,figsize=(15,15))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## Observation drawn from above wordcloud:
### Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency, this implies Maximum number of order are placed from these restaurants which are highlighted in the wordcloud.
